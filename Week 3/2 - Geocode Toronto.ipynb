* Merge both dataframes using pd.merge function, on PostalCode column

```python
import pandas as pd

#df = pd.read_csv('https://cocl.us/Geospatial_data')
#df.to_csv('toronto_latlon.csv', index = None)

latlong = pd.read_csv('toronto_latlon.csv')
latlong.rename(columns={"Postal Code": "PostalCode"}, inplace=True)

toronto = pd.read_csv('toronto_neig.csv')

result = pd.merge(toronto, latlong, on='PostalCode')
result
```
